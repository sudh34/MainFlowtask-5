```python
# 1. How many people have heart disease and how many doesn't have heart disease 

df.target.value_counts()
```

```python
#plotting the bar chart 
df.target.value_counts().plot(kind='bar',color = ['orchid','salmon'])
plt.title('Heart disease values')
plt.xlabel("1=heart disease , 0 = no heart disease ")
plt.ylabel("amount")
```

```python
#2.people of  which sex have most heart disease

pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data = df,hue = 'sex')
plt.title('heart disease frequency for sex')
plt.xlabel("0=no heart disease , 1= heart disease")
```

```python
#3.people of which sex have which type of chest pain 
df.cp.value_counts()
```

```python
#plotting a bar chart 
df.cp.value_counts().plot(kind='bar',color = ['salmon','lightskyblue','springgreen','khaki'])
plt.title('chest pain types vs count')
```

```python
pd.crosstab(df.sex,df.cp).plot(kind = 'bar', color = ['coral','lightskyblue','plum','khaki'])
plt.title('types of chest pain for sex')
plt.xlabel("0=female,1=male")
```

```python
#4.people of which chest pain have pron to have heart disease
pd.crosstab(df.cp,df.target)
```

```python
sns.countplot(x='cp',data=df,hue='target')
```

```python
#5.people of which age have most heart disease 
pd.crosstab(df.age,df.target)
```

```python
pd.crosstab(df.age,df.target).plot(kind = 'bar')
plt.title('age vs Heart disease')
plt.xlabel("0=female,1=male")
```

```python
#6.people of which fbs have high risk of having a heart disease"
pd.crosstab(df.fbs,df.target)
```

```python
pd.crosstab(df.fbs,df.target).plot(kind = 'bar')
plt.title('fbs having heart disease')
plt.xlabel("0=female,1=male")
```

```python
#7.people of which cholestrol have high risk of heart disease
pd.crosstab(df.oldpeak,df.target)
```

```python
pd.crosstab(df.oldpeak,df.target).plot(kind = 'bar')
plt.title('oldpeak vs  heart disease')
plt.xlabel("0=female,1=male")
```
